# Selenium

```python
from selenium import webdriver
import chromedriver_autoinstaller
import numpy as np
import pandas as pd
```

```python
#install browser such as chrome or edge
chromedriver_autoinstaller.install()

#use the installed browser for selection
browser = webdriver.Chrome()
```

```python
# it is isimilar to request.get
# (simply to get the website that we're tryin to scrape)

browser.get("https://arkansasrazorbacks.com/sport/m-basebl/stats/")
```

```python
from bs4 import BeautifulSoup #untuk merapihkan data yang diambi agar lebih rapi
```

```python
#make it beautiful html
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
soup.find_all('tbody', class_='row-hover')
```

```python
year = soup.find_all('td', class_='column-1')
```

```python
data_year = []
for each_year in year:
    data_year.append(each_year.text)
```

```python
data_year
```

```python
coach = soup.find_all('td', class_='column-2')
```

```python
data_coach = []
for each_coach in coach:
    data_coach.append(each_coach.text)
```

```python
data_coach
```

```python
record = soup.find_all('td', class_='column-3')
data_record = []
for each_record in record:
    data_record.append(each_record.text)
```

```python
data_record
```

```python
conf = soup.find_all('td', class_='column-4')
data_conf = []
for each_conf in conf:
    data_conf.append(each_conf.text)
```

```python
data_conf
```

```python
df = {
    'year':data_year,
    'coach':data_coach,
    'record':data_record,
    'conf':data_conf
}
```

```python
pd.DataFrame(df)
```

```python
pd.DataFrame(df).to_excel('Arkansas Razorbacks Stats.xlsx')
```
